# ZDRUŽI TABELE IN SHRANI V CSV

# 1. Naložimo vse csv datoteke

Ker nimajo vsa stanovanja na spletni strani podanega leta izgradnje in nadstropja, imam podatke o stanovanjih pobrane v štiri različne csv datoteke. Kar veže leto oz nadstopje stanovanja k pravemu stanovanju, je id. V tej datoteki bom najprej naložila vse štiri csv datoteke, nato bom naredila tabelo z indexom id ter ustreznimi leti in nadstropji ter jo na koncu združila z tabelo stanovanj, ki vsebuje vse ostale podatke. Ker ima vsak oglas enoličen id, se bosta tabeli združili pravilno. Na koncu bom dobljeno tabelo shranila v csv, da jo bom lahko uporabila za obdelavo podatkov. 

```python
import pandas as pd
stanovanja_leto = pd.read_csv('oglasi_leto1.csv', index_col = 'id', encoding = 'cp1250')
stanovanja_leto
```

```python
stanovanja_nadstropje = pd.read_csv('oglasi_nadstropje1.csv', index_col ='id', encoding = 'cp1250')
stanovanja_nadstropje
```

```python
stanovanja_nadstropje_leto = pd.read_csv('oglasi_nadstropje_leto1.csv',index_col = 'id', encoding = 'cp1250')
stanovanja_nadstropje_leto
```

```python
stanovanja = pd.read_csv('oglasi1.csv', index_col = 'id',  encoding = 'cp1250')
stanovanja
```

# 2. Združimo tabele

```python
tabela = stanovanja_nadstropje_leto.append([stanovanja_nadstropje, stanovanja_leto])
tabela
```

```python
result = pd.concat([stanovanja, tabela], axis=1, join_axes=[stanovanja.index])
result
```

# 3. Shranimo v csv

```python
result.to_csv('urejeni_oglasi1.csv')
```
