```python
enote_bezigrad = ['BEŽIGRAD', 'ČRNUČE', 'SAVSKO NASELJE', 'SAVLJE', 'BS 3', 'BEŽIGRAJSKI DVOR','ČRNUŠKA GMAJNA', 'ŽUPANČIČEVA JAMA', 'ZUPANČIČEVA JAMA', 'BRINJE', 'STOŽICE', 'TOMAČEVO']
frames_bezigrad = []
for i in enote_bezigrad:
    bezigrad = stanovanja[stanovanja['naslov'].str.contains(i)]
    frames_bezigrad.append(bezigrad)
result_bezigrad = pd.concat(frames_bezigrad)
result_bezigrad1 = result_bezigrad.drop_duplicates(subset = 'id')
result_bezigrad1
```

```python
stanovanja1 = stanovanja.append(result_bezigrad1).drop_duplicates(keep = False)
stanovanja1
```

```python
enote_siska = ['TACEN', 'ŠIŠKA', 'DRAVLJE', 'ŠENTVID', 'KOSEZE', 'PODUTIK', 'ŠMARTNO', 'PRŽAN']
frames_siska = []
for i in enote_siska:
    siska = stanovanja1[stanovanja1['naslov'].str.contains(i)]
    frames_siska.append(siska)
result_siska = pd.concat(frames_siska)
result_siska1 = result_siska.drop_duplicates(subset= 'id')
result_siska1
```

```python
stanovanja2 = pd.concat([stanovanja1, result_siska1]).drop_duplicates(keep = False)
stanovanja2
```

```python
enote_center = ['CENTER', 'VODMAT', 'STARA LJ', 'TABOR', 'POLJANE']
frames_center = []
for i in enote_center:
    center = stanovanja2[stanovanja2['naslov'].str.contains(i)]
    frames_center.append(center)
result_center = pd.concat(frames_center)
result_center1 = result_center.drop_duplicates(subset = 'id')
result_center1

```

```python
stanovanja3 = pd.concat([stanovanja2, result_center1]).drop_duplicates(keep = False)
stanovanja3
```
